### (9) Write a function takes a list value as an argument and returns a string with all the items separated by comma and a space, with *and* inserted before the last item.

sample output:   
If you enter a list `['Mathematics', 'computer', programming']`, then you should get `Mathematics, computer, and programming`.

```python
def list_to_string(spam):
    """
    Parameters
    ----------
    spam: list
        The input list.
    Returns
    -------
    string: str
        The ouput string.
    """
    # coding your answer here
    string = ""
    for i in range(len(spam)):
        if i != len(spam) - 1:
            string = string + spam[i] + ", "
        else:
            string = string + "and " + spam[i]
    return(string)
```
